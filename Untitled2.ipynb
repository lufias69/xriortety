```python
link_data = "52 3 Form Quisoner Penelitian Sri Hartati (Jawaban) (1).xlsx"

data = pd.read_excel(link_data, sheet_name="Sheet1")
data.head(1)
try:
    indes = data.nama
    data.set_index(indes, inplace=True)
    data = data.transpose()
    data = data.drop('nama')
    data.head()
except:
    pass
```
